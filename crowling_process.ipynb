```python
from selenium import webdriver                                                              
from selenium.webdriver.common.keys import Keys                               
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from time import sleep
import pandas as pd

```

```python
# a auxiliary function to check if the element that I am looking for is exists:
Driver = webdriver.Chrome()

def Is_exists(XPath):
    try:
        Driver.find_element(By.XPATH, XPath)
    except:
        return False
    return True
```

```python

#Openning a chrome tab via the url of the website and take all the urls of all the cars:

Driver = webdriver.Chrome()
links = []

for page in range (1, 380):

    URL = 'https://carwiz.co.il/used-cars?page={}'.format(page)
    Driver.get(URL)

    yearXpath = """//*[@id="main"]/div/div/div[4]/div[2]/div[{}]/div/div/div[1]/div[2]/div[3]/div/div[1]/p[2]"""
    itemXpath = """//*[@id="main"]/div/div/div[4]/div[2]/div[{}]"""
    URLXpath = """//*[@id="main"]/div/div/div[4]/div[2]/div[{}]/div/div/div[2]/div/div[2]/a"""

    i = 1

    while(1):
        if Is_exists(itemXpath.format(i)) == True:
            if Is_exists(yearXpath.format(i)) == True:
                link = Driver.find_element(By.XPATH, URLXpath.format(i)).get_attribute("href")
                links.append(link)
                

            i += 1    
        else:
            break
      
 # save all the urls in dataframe and csv file:
    df = pd.DataFrame({'URL': links})
    print(df)

      
    page += 1  

df.to_csv("DataFrame.csv", index = False)
```

```python
# insert to the dataframe the columns of all the Properties that I need:

df.loc[:,"version"] = 0
df.loc[:,"km"] = 0
df.loc[:,"engine"] = 0
df.loc[:,"year"] = 0
df.loc[:,"gear"] = 0
df.loc[:,"color"] = 0
df.loc[:,"origin_own"] = 0
df.loc[:,"test_cost"] = 0
df.loc[:,"price"] = 0
df.loc[:,"scanned"] = 0

df
```

```python
unscanned = 0
i = 1
```

```python
# run on all the urls that I haven't taking yet and take from them all of the data that I need:

links = unscanned_df['URL']

for link in links:

    Driver.get(link)
    priceXpath = """//*[@id="main"]/div[2]/div[2]/div[1]/div[1]/div[1]/div[4]/div[1]/div[1]/div"""
    detailsXpath = """//*[@id="main"]/div[2]/div[2]/div[1]/div[1]/div[2]/div/div/div[1]/div/table/tbody"""
    sleep(1)
    
    try: 
        
        details_table = Driver.find_element(By.XPATH, detailsXpath)
        list_of_details = details_table.find_elements(By.TAG_NAME, 'tr')
    
        for detail in list_of_details:

            data_type = detail.find_element(By.XPATH, 'td[1]').text
            data_val = detail.find_element(By.XPATH, 'td[2]').text
                            
            if data_type == "גירסה":
                df.loc[i, "version"] = data_val
            elif data_type == "שנה":
                df.loc[i, "year"] = data_val    
            elif data_type == "מנוע":
                df.loc[i, "engine"] = data_val  
            elif data_type == "קילומטר נוכחי":
                df.loc[i, "km"] = data_val  
            elif data_type == "יד":
                df.loc[i, "num_of_hand"] = data_val  
            elif data_type == "תיבת הילוכים":
                df.loc[i, "gear"] = data_val  
            elif data_type == "צבע":
                df.loc[i, "color"] = data_val  
            elif data_type == "בעלות מקורית":
                df.loc[i, "origin_own"] = data_val  
            elif data_type == "אגרת רישוי שנתית":
                df.loc[i, "test_cost"] = data_val       

            df.loc[i, "price"] = Driver.find_element(By.XPATH, priceXpath).text

            df.loc[i, "scanned"] = 1   

            df.to_csv('DataFrame.csv')
    except:
        unscanned += 1         

    i += 1    

print(unscanned)
df
```
